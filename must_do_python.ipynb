```python
no = int(input('give a no'))
if no < 2:
    print("not a prime")
else:
    count = 1
for j in range(1,no//2+1):
    if(no%j == 0):
      count= count+1
if(count==2):
        print('is a prime')
else: print('not a prime')
```

```python
'''Write a program in Python to find prime factors of a given integer.
● Example:
○ Input: 28
○ Output: [2, 7]'''
```

```python
no = int(input('give a no'))
factors =[]
prime_factors = []

for i in range(1,no+1):
  if(no%i==0):
    factors.append(i)
print(factors)
for i in factors:
  count=1
  for j in range(1,i//2+1):
    if(i%j == 0):
      count= count+1
  if(count==2):
      prime_factors.append(i)

print(prime_factors)

```

```python
'''Write a program in Python to find if a given number is perfect or not.
● Example:
○ Input: 28
○ Output: True'''
```

```python
no = int(input('give a no'))
def perfect(no):
    factors =[]
    for i in range(1,no):
      if(no%i==0):
        factors.append(i)
    print(factors)
    total = sum(factors)
    return total == no
print(perfect(no))
```
